```python
import re
import numpy as np
import pandas as pd
```

```python
data_dir = "/".join((".", "data"))
data_source = "test_data_public_utf8.csv"
data_df = pd.read_csv(open(data_dir + '/' + data_source, 'r', encoding='utf8'), delimiter=',')
data_df.head(10)
# data_df.loc[(data_df["Class"] == 1)]
```

```python
pattern = re.compile("(?P<word1>「[^」]*」)")
for corpus in enumerate(data_df.loc[(data_df["Class"] == 1), "Description"]):
    print("=========================================================")
    result = re.findall(pattern, corpus[1])
    temp_string = ""
    for text in enumerate(result):
        temp_string = temp_string + text[1]
    print("[%03d]:%s" % (corpus[0], temp_string))
    
    #print("=========================================================")
    #print(re.sub(pattern, '', corpus[1], 1))
    #print(pattern.sub(r'\g<word1>', corpus[1]))
```
